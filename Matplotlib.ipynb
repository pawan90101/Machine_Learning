```python
#line graph using numpy and line plot
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0.0,2.0,0.01)
s = 1 + np.cos(2*np.pi*x)
#plt.plot(x,s,'g')#g for colour green
#plt.plot(x,s,'*')#* for * lines
plt.plot(x,s,'--')# -- for -- lines
plt.grid()
plt.xlabel('Time(T)')
plt.ylabel('Voltage(MV)')
plt.title('Cosine wave plot (Cos(x))')
plt.show()
```

```python
# supplot plot in Matplotlib
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0.0,5.0)
x2 = np.linspace(0.0,2.0)
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)
plt.subplot(2,1,1)# we take 2 rows one column #last 1 is for first position of the graph 
plt.plot(x1,y1,'o-')
plt.title('Subplot 1')
plt.xlabel('X1')
plt.ylabel('AMP(Y1)')
plt.subplot(2,1,2)#last 2 is acquiring the second poisson
plt.plot(x2,y2,'.-')
plt.xlabel('X2')
plt.ylabel('AMP(Y2)')
plt.title('SubPlot 2')
plt.show()
```
