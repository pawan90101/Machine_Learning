```python
#line graph using numpy and line plot
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0.0,2.0,0.01)
s = 1 + np.cos(2*np.pi*x)
#plt.plot(x,s,'g')#g for colour green
#plt.plot(x,s,'*')#* for * lines
plt.plot(x,s,'--')# -- for -- lines
plt.grid()
plt.xlabel('Time(T)')
plt.ylabel('Voltage(MV)')
plt.title('Cosine wave plot (Cos(x))')
plt.show()
```

```python
# supplot plot in Matplotlib
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0.0,5.0)#linspace stands for linear space 
x2 = np.linspace(0.0,2.0)
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)
plt.subplot(2,1,1)# we take 2 rows one column #last 1 is for first position of the graph 
plt.plot(x1,y1,'o-')
plt.title('Subplot 1')
plt.xlabel('X1')
plt.ylabel('AMP(Y1)')
plt.subplot(2,1,2)#last 2 is acquiring the second poisson
plt.plot(x2,y2,'.-')
plt.xlabel('X2')
plt.ylabel('AMP(Y2)')
plt.title('SubPlot 2')
plt.show()
```

```python
#bar graph or bar plot
#bar plot use when we try meause only one side.
import matplotlib.pyplot as plt
import numpy as np
x = [1,2,3,4,5]
y = [10,24,36,40,5]
tick_lables = ['One','Two','three','four','five']
plt.bar(x,y,tick_label = tick_lables, width=0.7,color = ['g','r','b'])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Graph')
plt.show()
```

```python
#histogram plot use for frequency measurement for a variable.
#for example distribute age in a particular country.
import matplotlib.pyplot as plt
import numpy as np
ages = [2,14,18,25,30,40,50,60]
range = (0,70)
bins = 10
plt.hist(ages,bins,range,color='g',histtype='bar',rwidth=0.7)
plt.xlabel('Ages')
plt.ylabel('Bins')
plt.title('Histogram Plot')
plt.show()
```

```python
# Scatter Plot
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randint(0,100,10)
y = np.random.randint(0,100,10)
print(x)
print(y)
plt.scatter(x,y,label = 'stars',edgecolors='r',marker="*",s=30)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

```python
#Pie Chart when you show the result in percentage.
import matplotlib.pyplot as plt
import numpy as np
activities = ['eat','sleep','work','play']
slices = [3,7,8,6]
color = ['r','g','b','m']
plt.pie(slices,labels=activities,colors=color,startangle=90,shadow=True,explode=(0.2,0.1,0,0),autopct="%1.2f%%")
plt.legend()
plt.show()
```

```python
#plot mearging
import matplotlib.pyplot as plt
import  numpy as np

x = np.arange(0,2*np.pi,0.01)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1,label = "sin")
plt.plot(x,y2,label= 'cos')
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sin and Cosine')
plt.show()
```

```python
#Plot date from csv or text files
import matplotlib.pyplot as plt
import numpy as np
import csv

x = []
y = []
with open('sample.csv','r') as csvfile:
    plots = csv.reader(csvfile)
    for col in plots:
        x.append(col[0])
        y.append(col[1])
y.sort()
plt.plot(x,y,label='File')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Test Graph')
plt.legend()
plt.show()
```
